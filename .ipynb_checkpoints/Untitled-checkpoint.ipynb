```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import numpy as np
data = pd.read_json('planning.json')

#see some data information regarding each column type and statistical information about the numeric variables
#print(data.info())
#print(data.describe())
```

### Check requirements

First, I checked if the data meets all the requirements.

```python
#Check if all requisites are met
def checkReqs(df):

    """
    Parameters
    ----------
    df : TYPE - Pandas Dataframe
         Prints a line checking if each of the individual requisites are met
    Returns 
    -------
    A boolean True if all the required requisites are met in the given dataset, False otherwise
    """
    u_id = df['id'].is_unique
    print("Ids are all unique: "+ str(u_id))
    u_oid = df['originalId'].is_unique
    print("Original Ids are all unique: "+ str(u_oid))
    
    id_c = df['id'].notnull().any()
    print("There are no null Ids: " + str(id_c))
    
    o_id = df['originalId'].notnull().any()
    print("There are no null Original Ids: " + str(o_id))
    
    o_units = df['operatingUnit'].notnull().any()
    print("There are no null Operating Units: " + str(o_units))
    
    office_pc = df['officePostalCode'].notnull().any()
    print("There are no null Office Postal Codes: " + str(office_pc))
    
    t_hours = df['totalHours'].notnull().any()
    print("There are no null Total Hours: "+ str(t_hours))
    
    s_date = df['startDate'].notnull().any()
    print("There are no null Starting Dates: " + str(s_date))
    
    e_date = df['endDate'].notnull().any()
    print("There are no null Ending Dates: " + str(e_date))
    
    c_id = df['clientId'].notnull().any()
    print("There are no null Client Ids: " + str(c_id))
    
    return (u_id and u_oid and id_c and o_id and o_units and office_pc and t_hours and s_date and e_date and c_id)


print("All data requisites are met: " +str(checkReqs(data)) ) 
```

### Dates Overview

Starting the overview of the data, I looked at the distribution of the startDate and endDate to see if there was any date with an extreme value. It was hard to take any conclusions having the data grouped by day, the only date that got my attention was the 13th of August for the endDate, which had a value slightly higher than all the others. This made me wonder if there was any day of the week or month with a higher number of rows than the others. 

Studying the start date according to the day of the week, I noticed that the values were slightly lower on Sunday, Friday and Tuesday when compared to the other days of the week. For the end date, Monday was clearly the day of the week with a higher value.

On the month analysis, I noticed that both the start date and end date had a lower value for January and February when compared to the other months of the year. For the start date, all the other months seem to have a similar value, while for the end date, March and December seem to have a higher value than the other months.

```python
def datesOverview(data):
    df=data
    df['startDate'] = pd.to_datetime(pd.to_datetime(df['startDate'], format='%m/%d/%Y %H:%M %p').dt.strftime('%Y-%m-%d'))
    df['endDate'] = pd.to_datetime(pd.to_datetime(df['endDate'], format='%m/%d/%Y %H:%M %p').dt.strftime('%Y-%m-%d'))
        
    a = df['startDate'].value_counts().sort_index()
    x= a.index.tolist()
    y=a.tolist()
    plt.bar(x=x,height=y,color='blue')    
    plt.title("Start Date distribution")
    plt.show()
    plt.close()
    
    a = df['endDate'].value_counts().sort_index()
    x= a.index.tolist()
    y=a.tolist()
    plt.bar(x=x,height=y,color='blue')    
    plt.title("End Date distribution")
    plt.show()
    plt.close()
    
    df['startDate'] = pd.to_datetime(pd.to_datetime(df['startDate'], format='%m/%d/%Y %H:%M %p').dt.strftime('%Y-%m-%d'))
    df['endDate'] = pd.to_datetime(pd.to_datetime(df['endDate'], format='%m/%d/%Y %H:%M %p').dt.strftime('%Y-%m-%d'))    

    week_df = df.groupby(df['startDate'].dt.day_name(), as_index=False).size().sort_values('size')
    sns.barplot(data = week_df ,x="startDate",y="size",
                  dodge=False, ci=None)
    plt.xticks(rotation=15)
    plt.ylabel(None)
    plt.xlabel(None)
    plt.title("Start Date Distribution by Day of the Week")
    plt.show()
    plt.close()
    
    week_df = df.groupby(df['endDate'].dt.day_name(), as_index=False).size().sort_values('size')
    sns.barplot(data = week_df ,x="endDate",y="size",
                  dodge=False, ci=None)
    plt.xticks(rotation=15)
    plt.ylabel(None)
    plt.xlabel(None)
    plt.title("End Date Distribution by Day of the Week")
    plt.show()
    plt.close()
    
    month_df = df.groupby(df['startDate'].dt.month, as_index=False).size().sort_values('size')
    sns.barplot(data = month_df ,x="startDate",y="size",dodge=False, ci=None)
    plt.ylabel(None)
    plt.xlabel(None)
    plt.title("Start Date Distribution by Month of the Year")    
    plt.show()
    plt.close()
        
    month_df = df.groupby(df['endDate'].dt.month, as_index=False).size().sort_values('size')
    sns.barplot(data = month_df ,x="endDate",y="size",dodge=False, ci=None)
    plt.ylabel(None)
    plt.xlabel(None)
    plt.title("End Date Distribution by Month of the Year")    
    plt.show()
    plt.close()  
datesOverview(data)
```

### Statistics regarding the booking grade attribute

In this section I made a more in-depth analysis of the booking grade attribute.

The first thing we can see in this section is the relation of the booking grade attribute with the total number of hours. Even though all booking grades seem to have a similar total number of hours, the Interns seem to have a slightly higher average.

In the top 5 cities with more rows in the dataset, the main booking grades can be seen in the second plot of this section. I opted to leave the rows with no booking grade outside of this plot as it would significantly diminish the others and they would become hard to see. I noticed that there are no interns or senior associates in Mei√üen and no lead associates in Pirmasens and even though Gro√üenhain has more rows than Gerolzhofen, in the latter, there are more talents with a booking grade.

Regarding the booking grades in the top 5 industries with the most rows in the dataset, in the third plot of this section, we can see that the music industry is the one with the highest number of interns and lowest number of Managers and Lead Associates. On the other hand, the Film industry, is the one with the lowest number of interns and highest number of Senior Managers, together with the Textile industry.

The fourth plot of this section shows the relation of the mean duration (end date - start date) according to the booking grade. This duration is very similar for every booking grade. Nevertheless, it is slightly lower when it is an associate, manager or when there is no booking grade.

The last plot of this section is a heatmap that shows the number of rows that each of the booking grades has in each of the operating units. Since the rows with no booking grade are significantly more than all the others, the heatmap colors remained very similar for all the other booking grades. It is visible that the Operating Unit 7 is the one with the highest number of Interns, while Operating Unit 10 is the one with the lowest number of Interns.

```python
def bookingGradeAnalysis(data):
    """
        Parameters
    ----------
    data : TYPE Pandas Dataframe
        DESCRIPTION.
        Creates plots to help analysing the booking Grade
    Returns
    -------
    None.

    """
    df = data[['bookingGrade','totalHours']] # Get only the bookingGrade and totalHours columns
    df.loc[df["bookingGrade"] == '','bookingGrade'] = "No booking Grade"
    sns.boxplot(x="bookingGrade", y="totalHours",
                  palette=['blue','green','yellow','orange','red','pink','darkviolet'], data=df).set(
                      title='Number of Total Hours according to the Booking Grade')
    plt.xticks(rotation=20)
    plt.xlabel(None)
    plt.show()
    plt.close()
    
    
    df = data[['bookingGrade','officeCity']] 
    top5_cities = df['officeCity'].value_counts().head(5)
    print("The city with the most talents is: " +str(top5_cities.index[0])+ " with "+str(top5_cities[0])+" talents.")
    print("The second city with the most talents is: " +str(top5_cities.index[1])+ " with "+str(top5_cities[1])+" talents.")
    print("The third city with the most talents is: " +str(top5_cities.index[2])+ " with "+str(top5_cities[2])+" talents.")
    print("The fourth city with the most talents is: " +str(top5_cities.index[3])+ " with "+str(top5_cities[3])+" talents.")
    print("The fifth city with the most talents is: " +str(top5_cities.index[4])+ " with "+str(top5_cities[4])+" talents.")
    df=df.loc[df['officeCity'].isin(top5_cities.index)]
    index_t = data['bookingGrade'].value_counts().index.tolist()[1:]
    df=df.loc[df['bookingGrade'].isin(index_t)]
    ax=sns.histplot(binwidth=0.5, x="officeCity", hue="bookingGrade", data=df, stat="count", multiple="stack")
    plt.xlabel(None)
    plt.title("Booking Grades in the top5 cities")
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
    plt.show()
    plt.close()
    
    df = data[['bookingGrade','industry']] 
    top5_industries = df['industry'].value_counts().head(5)
    print("The industry with the most talents is: " +str(top5_industries.index[0])+ " with "+str(top5_industries[0])+" talents.")
    print("The second industry with the most talents is: " +str(top5_industries.index[1])+ " with "+str(top5_industries[1])+" talents.")
    print("The third industry with the most talents is: " +str(top5_industries.index[2])+ " with "+str(top5_industries[2])+" talents.")
    print("The fourth industry with the most talents is: " +str(top5_industries.index[3])+ " with "+str(top5_industries[3])+" talents.")
    print("The fifth industry with the most talents is: " +str(top5_industries.index[4])+ " with "+str(top5_industries[4])+" talents.")
    df=df.loc[df['industry'].isin(top5_industries.index)]
    index_t = data['bookingGrade'].value_counts().index.tolist()[1:]
    df=df.loc[df['bookingGrade'].isin(index_t)]
    ax= sns.histplot(binwidth=0.5, x="industry", hue="bookingGrade", data=df, stat="count", multiple="stack")
    plt.xlabel(None)
    plt.title("Booking Grades in the top5 industries")
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
    plt.show()
    plt.close()
    
    
    df=data
    df['startDate'] = pd.to_datetime(pd.to_datetime(df['startDate'], format='%m/%d/%Y %H:%M %p').dt.strftime('%Y-%m-%d'))
    df['endDate'] = pd.to_datetime(pd.to_datetime(df['endDate'], format='%m/%d/%Y %H:%M %p').dt.strftime('%Y-%m-%d'))
    df['Duration'] = (df['endDate'] - df['startDate']).dt.days
    df.loc[df["bookingGrade"] == '','bookingGrade'] = "No booking Grade"
    df = df.groupby('bookingGrade').mean().reset_index()
    df= df.sort_values('Duration')
    plt.bar(range(len(df)), df['Duration'], color=['blue','green','palevioletred','orange','red','darkviolet'])
    plt.xticks(range(len(df)),df['bookingGrade'], rotation=25)
    plt.ylabel('Mean Duration (in Days)')
    plt.title("Mean Duration in Days according to the Booking Grade")
    plt.show()
    plt.close()
    
    
    
    df = data[['bookingGrade','operatingUnit']]
    
    for row,col in df.iterrows():
        df.at[row,'operatingUnit'] = int(col['operatingUnit'].split(" ")[2])  # get just the int from the operating units in order to sort it correctly, string sort would put 10 before 2
        
    df = df.sort_values('operatingUnit')
    df = df.groupby(['bookingGrade','operatingUnit']).size().to_frame('size').reset_index()
    df['operatingUnit']=df['operatingUnit'].apply(str)
    for row,col in df.iterrows():
        df.at[row,'operatingUnit'] = "Operating Unit " + str(col['operatingUnit'])
        
        
    df.loc[df["bookingGrade"] == '','bookingGrade'] = "No booking Grade"
    bookingGrades = df['bookingGrade'].unique()
    operatingUnits = df['operatingUnit'].unique()
    sizes = df['size']
    l_sizes = [[]]
    for i in range(0,7):
        aux = []
        for j in range(0,10):
            result = j
            result += i*10
            aux.append(sizes[result])
        l_sizes.append(aux)
    l_sizes=l_sizes[1:]

    fig, ax = plt.subplots()
    im = ax.imshow(l_sizes)
    ax.set_xticks(np.arange(len(operatingUnits)))
    ax.set_yticks(np.arange(len(bookingGrades)))
    ax.set_xticklabels(operatingUnits)
    ax.set_yticklabels(bookingGrades)
    
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")


    for i in range(len(bookingGrades)):
        for j in range(len(operatingUnits)):
            text = ax.text(j, i, l_sizes[i][j],ha="center", va="center", color="w")
    ax.set_title("Number of talents in each Operating Unit")
    fig.tight_layout()
    plt.show()
    plt.close()
    
bookingGradeAnalysis(data)
```

### Statistics regarding the Skills

To conclude, here we have two pie charts, one regarding the required skills and one regarding the optional skills, where we can easily see the percentage of each skill.

```python
def skillsAnalysis(data):   
    
    a = data['requiredSkills']
    r_skills= []
    for x in a:
            for i in x: 
                r_skills.append( i['name'])
    rSkills = Counter(r_skills)
    l_keys = list(rSkills.keys())
    l_values = list(rSkills.values())
    plt.pie(l_values, labels = l_keys,autopct='%1.1f%%')
    plt.title("Required Skills Pie Chart")
    plt.show() 
    plt.close()
    
    b = data['optionalSkills']
    r_skills= []
    for x in b:
            for i in x: 
                r_skills.append( i['name'])
    rSkills = Counter(r_skills)
    l_keys = list(rSkills.keys())
    l_values = list(rSkills.values())
    plt.pie(l_values, labels = l_keys,autopct='%1.1f%%')
    plt.title("Optional Skills Pie Chart")
    plt.show() 
    plt.close()
skillsAnalysis(data)
```
